Укажите возраст (Age) самого молодого игрока, у которого хладнокровие (Composure) и реакция (Reactions) превышают 90% от максимального значения, представленного в датасете.

Насколько средняя реакция (Reactions) самых взрослых игроков больше средней реакции самых молодых игроков?

```python
data1 = df[df.Age == df.Age.max()]
data2 = df[df.Age == df.Age.min()] 
data3 = data1.Reactions.mean()- data2.Reactions.mean()
display(data3)
```

Во сколько раз средняя зарплата (Wage) голкипера (Position, GK) с максимальным значением показателя " Рефлексы" (GKReflexes) выше средней зарплаты голкипера с максимальным значением показателя "Владение мячом" (GKHandling)?

```python
data1 = df[(df.Position == 'GK')&(df.GKReflexes == df.GKReflexes.max())]
data2 = df[(df.Position == 'GK')&(df.GKHandling == df.GKHandling.max())]
data3 = data1.Wage.mean() / data2.Wage.mean()
display(data3)
```

Во сколько раз средняя сила удара (ShotPower) самых агрессивных игроков (игроков с максимальным значением показателя "Агрессивность" (Aggression)) выше силы удара игроков с минимальной агрессией?

```python
data1 = df[df.Aggression == df.Aggression.max()]
data2 = df[df.Aggression == df.Aggression.min()]
data3 = data1.ShotPower.mean() / data2.ShotPower.mean()
display(data3)
```
